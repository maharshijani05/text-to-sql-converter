```python
from datasets import load_dataset
import pandas as pd
import re

# Load Spider dataset
dataset = load_dataset("spider")

# Convert to Pandas DataFrames
train_df = pd.DataFrame(dataset["train"])
validation_df = pd.DataFrame(dataset["validation"])

# Function to clean natural language queries
def clean_text(text):
    text = text.lower().strip()  # Lowercase and strip whitespace
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special characters
    return text

# Function to normalize SQL queries
def normalize_sql(query):
    query = query.lower().strip()  # Lowercase and strip whitespace
    query = re.sub(r"\s+", " ", query)  # Remove extra spaces
    return query

# Apply preprocessing
train_df["question"] = train_df["question"].apply(clean_text)
train_df["query"] = train_df["query"].apply(normalize_sql)

validation_df["question"] = validation_df["question"].apply(clean_text)
validation_df["query"] = validation_df["query"].apply(normalize_sql)

# Remove missing values
train_df.dropna(inplace=True)
validation_df.dropna(inplace=True)

# Save cleaned data
train_df.to_csv("spider_train_cleaned.csv", index=False)
validation_df.to_csv("spider_validation_cleaned.csv", index=False)

print("✅ Preprocessing complete! Cleaned data saved as spider_train_cleaned.csv & spider_validation_cleaned.csv.")
```

```python
from datasets import load_dataset
import pandas as pd
import re
import torch
from transformers import AutoTokenizer

# Load Spider dataset
dataset = load_dataset("spider")

# Convert to Pandas DataFrames
train_df = pd.DataFrame(dataset["train"])
validation_df = pd.DataFrame(dataset["validation"])

# Function to clean text
def clean_text(text):
    text = text.lower().strip()
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    return text

# Function to normalize SQL queries
def normalize_sql(query):
    query = query.lower().strip()
    query = re.sub(r"\s+", " ", query)
    return query

# Apply preprocessing
train_df["question"] = train_df["question"].apply(clean_text)
train_df["query"] = train_df["query"].apply(normalize_sql)

validation_df["question"] = validation_df["question"].apply(clean_text)
validation_df["query"] = validation_df["query"].apply(normalize_sql)

# Load tokenizer (BERT or T5)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize text queries
train_questions = tokenizer(train_df["question"].tolist(), padding=True, truncation=True, return_tensors="pt")
val_questions = tokenizer(validation_df["question"].tolist(), padding=True, truncation=True, return_tensors="pt")

# Tokenize SQL queries
train_queries = tokenizer(train_df["query"].tolist(), padding=True, truncation=True, return_tensors="pt")
val_queries = tokenizer(validation_df["query"].tolist(), padding=True, truncation=True, return_tensors="pt")

# Save tokenized tensors
torch.save(train_questions, "train_questions.pt")
torch.save(train_queries, "train_queries.pt")
torch.save(val_questions, "val_questions.pt")
torch.save(val_queries, "val_queries.pt")

print("✅ Feature engineering complete! Tokenized data saved as PyTorch tensors.")
```

```python
import torch
from datasets import load_dataset
from transformers import T5ForConditionalGeneration, T5Tokenizer, Trainer, TrainingArguments

# Load Spider dataset
dataset = load_dataset("spider")

# Load T5 tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-large")

# Function to preprocess data for training
def preprocess_data(examples):
    inputs = ["translate to SQL: " + q for q in examples["question"]]
    targets = examples["query"]

    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Tokenize dataset
train_data = dataset["train"].map(preprocess_data, batched=True)
val_data = dataset["validation"].map(preprocess_data, batched=True)

# Load T5 model
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Training arguments
training_args = TrainingArguments(
    output_dir="./t5_sql_model",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-4,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=5,
    weight_decay=0.01,
    save_total_limit=2,
    push_to_hub=False,
    logging_dir="./logs",
    logging_steps=10,
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
)

# Train the model
trainer.train()

# Save model
model.save_pretrained("t5_sql_model")
tokenizer.save_pretrained("t5_sql_model")

print("✅ Training Complete! Model saved as 't5_sql_model'.")
```

```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the trained model & tokenizer
model_path = "t5_sql_model"  # Change this if your model is saved elsewhere
model = T5ForConditionalGeneration.from_pretrained(model_path)
tokenizer = T5Tokenizer.from_pretrained(model_path)

# Function to generate SQL from natural language questions
def generate_sql(question):
    input_text = "translate to SQL: " + question
    input_ids = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True).input_ids

    with torch.no_grad():
        output_ids = model.generate(input_ids, max_length=128)

    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

# User Input Loop
while True:
    user_question = input("\n🔹 Enter a question (or type 'exit' to quit): ")
    if user_question.lower() == "exit":
        print("🚀 Exiting program. Have a great day!")
        break

    predicted_sql = generate_sql(user_question)
    print(f"\n✅ Generated SQL Query: {predicted_sql}\n")
```
